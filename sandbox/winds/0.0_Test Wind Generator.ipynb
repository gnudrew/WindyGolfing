# Test Wind Generator

## Oscillatory

We can adjust the amplitude, frequency, and phase_offset vectors (x,y,z) to shape how the wind velocity oscillates. In the example configurations below, the parametric plots visualize how the wind velocity vector evolves over time. No matter how complex the pattern, this generator is fundamentally deterministic and thus should be predictable by a skilled agent.

```python
# oscillate along X-axis
params = {
    'base_velocity': np.array([0,0,0]),
    'amplitude': np.array([1,0,0]),
    'frequency': np.array([1,1,1]),
    'phase_offset': np.array([1,0,0]),
    'dt': .001,
}
G1 = OscillatoryGenerator(params)
G1.gen(10)

G1.plot3d()
G1.plotx()
G1.ploty()
G1.plotz()
```

```python
# oscillate in XY-plane (Circle)
params = {
    'base_velocity': np.array([0,0,0]),
    'amplitude': np.array([1,1,0]),
    'frequency': np.array([1,1,1]),
    'phase_offset': np.array([0,np.pi/2,0]),
    'dt': .001,
}
G1 = OscillatoryGenerator(params)
G1.gen(10)

G1.plot3d()
G1.plotx()
G1.ploty()
G1.plotz()
```

## Lorenz

The classic butterly shape. A tiny perturbation to `base_velocity`, say `v = [1,1,1]` to `v = [1.1,1,1]` quickly diverges to a very different trajectory.

```python
params = {
    'base_velocity': np.array([1,1,1]),
    'rho': 28,
    'sigma': 10,
    'beta': 8/3,
    'dt': .001
}
G2 = LorenzGenerator(params)
G2.gen(30)

G2.plot3d()
```

```python
params = {
    'base_velocity': np.array([1.1,1,1]),
    'rho': 28,
    'sigma': 10,
    'beta': 8/3,
    'dt': .001
}
G2 = LorenzGenerator(params)
G2.gen(30)

G2.plot3d()
```
