## Alternatively, generate a Windspacetime on the fly

```python
# oscillate along x-axis only.
timestep = .01
params = {
    'base_velocity': np.array([0,0,0]),
    'amplitude': np.array([1,0,0]),
    'frequency': np.array([1,1,1]),
    'phase_offset': np.array([0,0,0]),
    'dt': timestep,
}
G = OscillatoryGenerator(params)
arr_wind_v = G.gen(10)
G.plot3d()
G.plotx()
```

```python
arr_wind_v
```

## Run Simulation

### Set starting conditions

```python
t_initial = 0
p_initial = np.array([0,0,10])
v_initial = np.array([0,0,0])
arr_windspacetime = arr_wind_v
timestep = timestep
```

### Run simulator

```python
runner = SimTrialRunner(t_initial, p_initial, v_initial, arr_windspacetime, timestep)
runner.run()
```

```python
runner.plotx()
runner.ploty()
runner.plotz()
```
